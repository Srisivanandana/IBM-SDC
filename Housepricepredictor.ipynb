```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Sample Dataset
data = {
    'Size (sqft)': [1000, 1500, 2000, 2500, 3000],
    'Price (Lakhs)': [50, 65, 80, 100, 120]
}

df = pd.DataFrame(data)

X = df[['Size (sqft)']]  # Feature
y = df['Price (Lakhs)']  # Target

# Splitting dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Training the model
model = LinearRegression()
model.fit(X_train, y_train)

# Getting input from user
size = float(input("Enter the size of the house in sqft: "))

# Predicting price
predicted_price = model.predict([[size]])

print(f"Predicted price for a house of {size} sqft is: {predicted_price[0]:.2f} Lakhs")
```
