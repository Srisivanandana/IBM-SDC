```python
# Step 1: Install dependencies (run this first in Colab/Jupyter)
!pip install -q google-generativeai

# Step 2: Import required libraries
import google.generativeai as genai

# Step 3: Configure Gemini with your API key
GOOGLE_API_KEY = "AIzaSyDuSZxPH--lMe0rNoj-afyeKezzYcnF8Bk"  # Replace with your actual API key
genai.configure(api_key=GOOGLE_API_KEY)

# Step 4: Initialize the Gemini model (using 1.5 Flash)
model = genai.GenerativeModel('gemini-1.5-flash')

# Step 5: Define the translation function
def translate_text(text: str, target_language: str) -> str:
    """
    Translates text to the specified target language using Gemini.

    Args:
        text: The text to translate
        target_language: The target language (e.g., "French", "Spanish")

    Returns:
        The translated text
    """
    prompt = f"""
    You are a professional translator.
    Translate the following text to {target_language}.
    Maintain the original meaning and tone.

    Text: {text}

    Provide only the translation without additional commentary.
    """

    try:
        response = model.generate_content(prompt)
        return response.text.strip()
    except Exception as e:
        return f"Translation error: {str(e)}"

# Step 6: Example usage
if __name__ == "__main__":
    # Example translation
    original_text = "Good morning! How are you today?"
    target_language = "French"

    translated_text = translate_text(original_text, target_language)

    print(f"üî§ Original: {original_text}")
    print(f"üåê Translated ({target_language}): {translated_text}")
```
